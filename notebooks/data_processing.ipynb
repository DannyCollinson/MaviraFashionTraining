```python
import os

# The stage at which to stop the processing pipeline. If -1, run the entire
# pipeline. If 0, stop after cleaning the filenames. If 1, stop after resizing.
# If 2, stop after splitting. If 3, stop after normalizing the split images
stage = -1

# If starting from raw data, this is an absolute path or a relative path from
# this notebook to the root directory of the raw dataset as a string.
# Ignored if not starting from raw data
raw = "../data/test/classifier4/"
# raw = "../data/classifier/classifier226/"

# If starting from raw data, this is an absolute or relative path to where the
# resized images will be stored. If not starting from raw data, this is a
# path to the root directory of a dataset that has already been resized.
# Ignored if not resizing images or starting from resized data
resized = None

# If starting from raw or resized data, this is an absolute or relative path to
# where the split images will be stored. If starting with normalization, this
# is a path to the root directory of a dataset that has already been split into
# train/val/test sets. If only performing format conversion, this is a path to
# the root directory of a dataset that has already been normalized.
# Ignored if not splitting images or starting from split or normalized data
split = None

# If starting from raw, resized, split, or normalized data, this is an absolute
# or relative path to where the converted images will be stored.
# Ignored if not performing format conversion
converted = None

# The height that images will be resized to. Ignored if not resizing images
height = 224

# The width that images will be resized to. Ignored if not resizing images
width = 224

# The interpolation method used for resizing. Ignored if not resizing images
interp = "bicubic"

# Ratio of the dataset to use for training/validation/testing.
# Ignored if not splitting into train/val/test sets
ratios = "60/15/25"

# The normalization method used for the images
# Currently supported options are
# "zscore", "minmax", "pixelz", "minmaxplus" and "none"
# Ignored if not normalizing images
norm_method = "zscore"

# The file format to convert images to. Currently supported options are
# "jpg", "npy", "pt", and "none" (keep images as "pt" PyTorch tensors)
format = "npy"

# If converting to JPEG, the quality 0-100 of the images after compression.
# Ignored if not converting images to JPEG
quality = None

# Whether or not to remove any intermediate files created during processing
# (0 for no, 1 for yes). Ignored if not running the entire pipeline
cleanup = 0

# Make sure that you have verified all parameters before running this cell
run = os.system(
    "python ../scripts/run_data_processing.py "
    f"--raw={raw} --resized={resized} "
    f"--split={split} --converted={converted} "
    f"--height={height} --width={width} --interp={interp} "
    f"--ratios={ratios} --norm_method={norm_method} "
    f"--format={format} --quality={quality} "
    f"--cleanup={cleanup} "
    f"{stage}"
)
```
